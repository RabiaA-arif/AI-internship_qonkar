```python
from langchain_community.document_loaders import WebBaseLoader

loader=WebBaseLoader("https://www.guru.com/m/find/freelance-jobs/web-data-scraping/")
page_data=loader.load().pop().page_content
print(page_data)
```

```python
from langchain_core.prompts import PromptTemplate

prompt_extract=PromptTemplate.from_template(
    """
    ### Subject: getting job in company :
    hi rabia
    {page_data}
    ### INSTRUCTION:
    The scraped text from carreris page of a website .
    Your job is to write cold email form to this job postings and return in text format
    followings keys : 'role','exprience','skills' and 'description'
    
    ### VALID text (NO PERMEABLE):
    """
)
chain_extract=prompt_extract| llm
res=chain_extract.invoke(input={'page_data':page_data})
print(res.content)
```

```python
from langchain_core.output_parsers import JsonOutputParser

json_parser=JsonOutputParser()
json_res=json_parser.parse(res.content)
json_res
```

```python
import pandas as pd
df=pd.read_csv("companies.csv")
df
```

```python
import uuid
import chromadb


client=chromadb.PersistentClient('vectorstore')
collection=client.get_or_create_collection(name="portfolio")
if not collection.count():
    for _, row in df.iterrows():
        collection.add(documents=row["Company Name"],
        metadatas={"links":row["Portfolio Link"]},
        ids=[str(uuid.uuid4())])
```
