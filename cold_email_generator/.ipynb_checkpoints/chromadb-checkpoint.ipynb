```python
import chromadb
chroma_client = chromadb.Client()
collection = chroma_client.create_collection(name="my_collection")
```

```python
collection.add(
    documents=[
        "This  documnet is the about pakistan",
        "This document about china"
        
    
    ],
    ids=["ids1","ids2"]
)
```

```python
all_documents=collection.get()
all_documents
```

```python
document=collection.get(ids=["ids1"])
document
```

```python
results=collection.query(
    query_texts=["Query is about Biryani"],
    n_results=2
)
results
```
